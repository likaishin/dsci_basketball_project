# Classifying NBA Players into Performance Categories

## Introduction
The National Basketball Association (NBA) is professional basketball league comprising 30 teams across North America. The project will focus on the NBA regular season which runs from October to April with 82 games. Sports analytics are a driving force for analyzing and improving the performance of teams and individual players. These analytics are particularly useful for team management decision-making, specifically in matters of player scouting, drafting, and roster adjustments. It is crucial that the data is accurate given these decisions are also financially related.  However, providing equal opportunities for players to show off their skills are also an important determining factor.

The objective of the project is to analyze a player’s individual performance in a game based on their age, minutes played, and field goal attempts. Regression analysis will be used to predict the average number of points a player will score per game. The predictive question is: __Can one use an NBA player’s age, playing time, and field goal attempts to predict the number of points they will score per game on average?__ These predictors were chosen to reveal trends across games, offering valuable information into how players may optimize their statistical performance during crucial moments in a game. 

The selected dataset shows NBA player stats during the regular season from 2022-2023. The dataset was created by an international organization called Creative Commons (CC) that aims to supply the world with knowledge that serves the public interest. The dataset is updated consistently to maintain the accuracy of the statistics.

##  Analysis Methods and Results

```r
library(repr)
library(tidyverse)
library(tidymodels)
options(repr.matrix.max.rows = 10)
```

In order to wrangle the data we first have to load in all the necessary libraries in order to utilize the functions they provide. We load in the tidyverse package as it is helpful for data visualization and has functions such as ggplot, and we load in tidymodels for the regression that will be done later on. 

First, we load the player data and convert the columns we need to numeric types, then assign it to an object named `player_data_numeric`. 

**Table 1: Raw Player Data**

```r
url <- "https://raw.githubusercontent.com/likaishin/dsci_basketball_project/main/NBA_Player_Stats_Regular.csv"
player_data <- read_delim(url, delim = ";")
player_data_numeric <- player_data |>
    mutate(Age = as.numeric(Age),
           PTS = as.numeric(PTS),
           FGA = as.numeric(FGA))
player_data_numeric
```

As this data frame is alread in tidy format, we do not have to wrangle the data. 

We filtered the players who are only playing the center position, as this is the position we selected to run our regression on. We also filtered the dataframe into centers who have played more than 20 games in order to increase the accuracy of our data and avoid any biases. Assign it to an object named `player_data_filter`.

**Table 2: Filtered Player Data**

```r
player_data_filter <- filter(player_data_numeric, Pos == "C" & G >= 20)
player_data_filter
```

Next, we select only the columns that we will utilize in the regression, the columns being age, MP (average minutes played per game), FGA (average field goal attempts per game), and PTS (average points scored per game). Assign it to an object named `player_data_select`.

**Table 3: Selected Player Data**

```r
player_data_select <- select(player_data_filter, Age, MP, FGA, PTS)
player_data_select 
```

We set seed to ensure reproducibility, and then split the dataset into the training and testing datasets, using 75% of the original data as the training data. We assign the training dataset to an object named `player_data_training`, and the testing dataset to an object named `player_data_testing`.

```r
set.seed(2020)
player_data_split <- initial_split(player_data_select, prop = 0.75, strata = PTS)
player_data_training <- training(player_data_split)
player_data_testing <- testing(player_data_split)
```

We conduct visualizations to explore the relationship between  MP (average minutes played per game), FGA (average field goal attempts per game), and PTS (average points scored per game) seperately.

**Figure 1: Average Playing Time vs. Average Points Per Game**

```r
play_plot_1 <- ggplot(player_data_training, aes(x = MP, y = PTS))+
                geom_point()+
                xlab("Average Playing Time (minutes)") +
                ylab("Average Points Per Game")+
                ggtitle("Average Playing Time vs. Average Points Per Game") +
                theme(text = element_text(size=15))
play_plot_1
```

**Figure 2: Average Field Goal Attempts Per Game vs. Average Points Per Game**

```r
play_plot_2 <- ggplot(player_data_training, aes(x = FGA, y = PTS))+
                geom_point()+
                xlab("Average Field Goal Attempts Per Game") +
                ylab("Average Points Per Game") +
                ggtitle("Average Field Goal Attempts Per Game vs. Average Points Per Game") +
                theme(text = element_text(size=12))
play_plot_2
```

**Figure 3: Age of Players vs. Average Points Per Game**

```r
play_plot_3 <- ggplot(player_data_training, aes(x = Age, y = PTS))+
                geom_point() +
                xlab("Age of Players") +
                ylab("Average Points Per Game of Players ") +
                ggtitle("Age of Players vs. Average Points Per Game") +
                theme(text = element_text(size=12))
play_plot_3
```

Due to the presence of multiple individuals across various age groups, directly visualizing the data based on age alone might not be as intuitive. Therefore, we utilized the `group_by` and `summarize` functions to compute the average PTS for each age group, and assigned it to an object called `player_data_training_mean`. Subsequently, we created a visualization depicting the distribution of average PTS for players in different age groups. 

**Figure 4: Age of Players vs. Average Points of Players in Different Age Group**

```r
player_data_training_mean <- player_data_training |>
                             group_by(Age) |>
                             summarize(mean_PTS = mean(PTS))
play_plot_4 <- ggplot(player_data_training_mean, aes(x = Age, y = mean_PTS))+
                geom_point() +
                xlab("Age of Players") +
                ylab("Average Points Per Game of Players in Different Age Group") +
                ggtitle("Age of Players vs. Average Points of Players in Different Age Group") +
                theme(text = element_text(size=12))
play_plot_4
```

First, we perform a liner regression. We create a recipe with the variables of interest (PTS, Age, MP and FGA) using the training dataset. Then, we create a workflow analysis with our model specification and recipe, and fit in the training dataset as well. Assign it to an object named `player_fit`.

```r
set.seed(1234)

player_spec <- linear_reg() |>
        set_engine("lm") |>
        set_mode("regression")

player_recipe <- recipe(PTS ~ Age + MP + FGA, data = player_data_training)

player_fit <- workflow() |>
    add_model(player_spec) |>
    add_recipe(player_recipe) |>
    fit(data = player_data_training)
player_fit
```

**Table 4: Model Evaluation Results on Testing Data (Linear Regression)**

```r
set.seed(1234)

player_fit_results <- player_fit |>
    predict(player_data_testing) |>
    bind_cols(player_data_testing) |>
    metrics(truth = PTS, estimate = .pred)
player_fit_results
```

According to our visualization of the data, the relationship between “PTS” and “Minutes”, and “PTS” and “FGA” are linear but the relationship with “Age” is not. Therefore, it would be appropriate to run a $k$-nn regression model onto our dataset and compare the RMSE as a measure of how well each algorithm fits the model based on the training set. 

First, we perform a cross-validation and choose the optimal $k$, and create a recipe to preprocess our data, store the recipe in an object named `player_knn_recipe`.

```r
set.seed(1234) 

player_knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |> 
       set_engine("kknn") |>
       set_mode("regression") 

player_knn_recipe <- recipe(PTS ~ Age + MP + FGA, data = player_data_training) |>
       step_scale(all_predictors()) |>
       step_center(all_predictors())

player_knn_recipe
```

Then we create the splits for cross-validation with *5 folds* using the `vfold_cv` function and create a workflow analysis to combine the model specification and recipe. 

```r
set.seed(1234) 

player_vfold <- vfold_cv(player_data_training, v = 5, strata = PTS)

player_knn_workflow <- workflow() |>
                  add_recipe(player_knn_recipe) |>
                  add_model(player_knn_spec) 
```

We use the RMSPE to find the best setting for $k$ from our workflow and test the values $k = 1, 2, 3, 4, ..., 10$. Then tune the workflow to test all the values of $k$ and *resamples* using the cross-validation data set, and collect the statistics from the model. Assign it to an object called `player_results`.

**Table 5: K-NN Regression Result**

```r
set.seed(2019)
install.packages("kknn")
gridvals <- tibble(neighbors = seq(from = 1, to = 10, by = 1))
player_results <- player_knn_workflow|>
            tune_grid(resamples = player_vfold, grid = gridvals)|>
            collect_metrics()
player_results
```

Next, we find the *minimum* RMSPE along with it's associated metrics such as the mean and standard error, to find the number of neighbors that will serve as our best $k$ value. 

**Table 6: K-NN Minimum Result**

```r
set.seed(2020) 

player_min <- player_results|>
    filter(.metric == "rmse") |>
    slice_min(n = 1, order_by = mean)
player_min
```

Finally, we use the $K$ value we obtained above and perform $k$-nn regression, assign the result to an object named `player_summary`.

**Table 7: Model Evaluation Results on Testing Data ($k$-nn Regression)**

```r
set.seed(1234) 

k_min <- player_min |>
          pull(neighbors)

player_best_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = k_min) |>
          set_engine("kknn") |>
          set_mode("regression")

player_best_fit <- workflow() |>
          add_recipe(player_knn_recipe) |>
          add_model(player_best_spec) |>
          fit(data = player_data_training)

player_summary <- player_best_fit |>
           predict(player_data_testing) |>
           bind_cols(player_data_testing) |>
           metrics(truth = PTS, estimate = .pred)
player_summary
```

After a comparison of the RMSE value for linear regression and $k$-nn regression, we find that linear regression yields a RMSE of 0.976, and $k$-nn regression yields a RMSE of 1.430. These values of the RMSE are in units of the response variable, in this case points. It is clear that the linear regression model does a better job of making a prediction, at least within our training set. Considering that in an NBA game points can only go up by 1, 2 or 3, this signifies that our algorithm does a good job of predicting within the training set, with a very low discrepancy between the actual value and the prediction value. 

**Table 8: Coefficients of Predictors**

```r
player_coeffs <- player_fit |>
             extract_fit_parsnip() |>
             tidy()
player_coeffs
```

By using the linear regression algorithm, and by using the `extract_fit_parsnip` function to find the coefficients of the predictors we can now create a formula with our predictors to make a prediction of average player points. The formula for the prediction is as follows : 

$Average\:Points\:Per\:Game = -0.463638 + 0.005035 * Player\:Age + 0.040020 * Average\:Minutes\:Player + 1.389284 * Average\:Field\:Goal\:Attempts$

We can also note that the value of the coefficients relates to how impactful that predictors is to the outcome of the prediction. Which shows that FGA is the primary contributor to a players average points made, and player age, as well as minutes played does not play as significant of a role. 

However, this only applies to our training and testing set, which is subject to change in the real world. Luckily, as the NBA season had just begun recently, we can apply our formula for the prediction to actual NBA players and determine whether our algorithm has actually done a good job in predicting average player points. As previously discussed one application of this prediction is to determine how well a rookie will perform in the NBA. As we chose the center position when producing our algorithm we will be choosing two rookie centers. The two rookies we will be selecting are Victor Wembanyana and Chet Holmgren, two very promising rookies who are also the front runners for the NBA rookie of the year award. We will be using their 2023/2024 regular season stats thus far to make a prediction of their average points per game. 

We will now load in the data set for the 2023/2024 NBA regular season, assigned it to an object named `NBA_player_data`.

**Table 9: Raw Player Data 23-24**

```r
url <- "https://raw.githubusercontent.com/likaishin/dsci_basketball_project/main/2023-2024%20NBA%20Player%20Stats%20-%20Regular.csv"
NBA_player <- read_delim(url, delim = ";")
NBA_player_data <- NBA_player |>
    mutate(Age = as.numeric(Age),
           PTS = as.numeric(PTS),
           FGA = as.numeric(FGA))
NBA_player_data
```

First, we want to filter the data set according to player names, only keeping the two rookies that we want to analyze. 

**Table 10: Filtered Player Data 23-24**

```r
NBA_player_filter <- filter(NBA_player_data, Player == "Victor Wembanyama" | Player == "Chet Holmgren")
NBA_player_filter
```

We then select only the predictor variables as we will only be operating on these values. 

**Table 11: Selected Player Data 23-24**

```r
NBA_player_select <- select(NBA_player_filter, Player, Age, MP, FGA, PTS)
NBA_player_select
```

Now we have the 2023/2024 stats for *Chet Holmgren* and *Victor Wembanyama*, we can use their stats to predict a value for the average points they will score this season and compare it to the actual value. 

First, we will do the calculations for *Chet Holmgren*.

```r
Chet_Holmgren_PTS <- -0.463638 + 0.005035 * 21 + 0.040020 * 30.3 + 1.389284 * 10.8
Chet_Holmgren_PTS
```

```r
Chet_Percent_Error <- (abs(Chet_Holmgren_PTS - 17.1)/ 17.1) * 100
Chet_Percent_Error
```

By applying the formula we generated with our linear regression model we obtained a prediction value for *Chet Holmgren*'s average points this season which turned out to be around 15.9 points. By comparing this value with the actual value for *Chet Holmgren*'s average points, which is 17.1, we can calculate a percent error which sums out to be 7.26 percent. 

Now, we will repeat the calculations for *Victor Wembanyama*. 

```r
Victor_Wembanyama_PTS <- -0.463638 + 0.005035 * 20 + 0.040020 * 30.2 + 1.389284 * 16.8
Victor_Wembanyama_PTS
```

```r
Victor_Percent_Error <- (abs(Victor_Wembanyama_PTS - 19.3)/ 19.3) * 100
Victor_Percent_Error
```

Applying the formula for *Victor Wembanyama* gave a value of 24.2 average points, in comparison to his actual average of 19.3, which gives a percent error of around 25.31 percent. 

It is interesting to note that for *Chet Holmgren* the algorithm under-predicted his average points, while in the case of *Victor Wembanyama* the algorithm over-predicted his points. One explanation for this lies within the data set itself. After filtering for only the two players we see that these are the average stats across only 13 games. When we ran the algorithm for linear regression we filtered the data for players who have only played above 20 games, meaning the 2023/2024 data set does not yet have enough information for an accurate prediction. Additionally, *Victor Wembanyama* has an average FGA of 16.8, in comparison to *Chet Holmgren's* 10.8, which is likely the reason *Wembanyama* was over-predicted, as FGA is the most impactful predictor to the prediction. 

## Discussion:

Based on the chosen predictors, this dataset focused on how a player’s amount of experienceplay time correlates to their performance, in this case how many points they score per game. **A clear relationship between the number of points scored per game on average by an individual and their age, minutes played, and field goal attempts were expected to be found.**

It was founded that there was **a positive relationship between the minutes an individual plays per game and the number of points they score**.  This relationship is intuitive, as more playing time offers more opportunities to score. There was also **a positive relationship between the number of points scored and the amount of field goal attempts made by the player**. This could be due to players with more attempts being more skilled or aggressive in scoring, or simply having more opportunities to shoot. Understanding these relationships can inform coaching decisions. For instance, if an individual did not follow this positive trend, it might benefit the team to substitute them. Age was another predictor that is useful to see if age possibly corresponds their points scored as an older age could mean the individual tires out quicker and therefore underperforms after a certain length of time.  It was founded that **age was not a big determining factor of the number of points scored**, meaning that teams do not have to worry about a player’s ability to perform based on it. This analysis can be very valuable for basketball teams in strategizing and making informed decisions about player rotations and game plans. It highlight's the importance of looking deeper into statistics to consider how various factors interplay to affect a player's performance on the court.

*These statistics could be used to not only determine which players might be underperforming, but if expanded one could analyze the importance of player positions along with these predictors.* For example, if a player played multiple positions, these statistics could determine which position would be best to increase the overall performance of the player and the team. Teams can also analyze these trends in their opponents, identifying potential weaknesses or fatigue patterns in key players. This data can play a crucial role in scouting and recruitment, as it can help scouts identify players who are best suited for specific roles within a team. Such as starters who can maintain high performance over extended minutes or bench players who can provide efficient scoring in limited time. With a data-driven approach, scouts can work towards eliminating biases based on age, background, or physical attributes, focusing more on measurable performance metrics.

*There are many future questions that our findings could lead to*, for example, does the effect of playing time on scoring differ in teams with a more balanced attack versus those reliant on one or two-star players? We could infer if balanced teams benefit from having multiple scoring threats, making it harder for opponents to defend and thereby maintaining scoring efficiency over longer playing times. In contrast seeing if star-driven teams face more targeted defence tactics (like double teaming) and seeing how this impacts the scoring efficiency of star players with increased playing time. We can also look into how coaching strategies and tactics influence a player's ability to maintain scoring efficiency with increased playing time or field goal attempts. Coaching strategies and tactics play an important role in influencing a player's scoring efficiency. For example, the type of offensive system the coach employs have a large impact on player scoring opportunities and efficiency. Certain systems may suit a player's style better than others, therefore enhancing the number of points scored.
